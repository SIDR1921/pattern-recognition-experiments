```python
import numpy as np
from scipy.spatial import distance
```

```python
def polygon_to_vector(points):
    points = np.array(points)
    return points.flatten()

def cosine_similarity(poly1,poly2):
    v1,v2 = polygon_to_vector(poly1), polygon_to_vector(poly2)
    return 1 - distance.cosine(v1,v2)
```

```python
def jaccard_similarity(poly1,poly2):
    set1 = set(map(tuple,poly1))
    set2 = set(map(tuple, poly2))

    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection/union if union!=0 else 0
```

```python
def manhattan_distance(poly1, poly2): 
    v1, v2 = polygon_to_vector(poly1), polygon_to_vector(poly2) 
    return distance.cityblock(v1, v2)
```

```python
n1 = int(input("Enter number of vertices for Polygon 1: "))
poly1 = [tuple(map(float, input(f"Vertex {i+1} (x y): ").split())) for i in range(n1)]
n2 = int(input("Enter number of vertices for Polygon 2: ")) 
poly2 = [tuple(map(float, input(f"Vertex {i+1} (x y): ").split())) for i in range(n2)]
print("\nResults:") 
print("Cosine Similarity:", cosine_similarity(poly1, poly2))
print("Jaccard Similarity:", jaccard_similarity(poly1, poly2))
print("Manhattan Distance:", manhattan_distance(poly1, poly2))
```
